```python
import numpy as np
import tensorflow as tf
import tensorflow_hub as hub
import keras
import random
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import re
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import nltk
nltk.download('wordnet')
import os
```

```python
train_data, validation_data, test_data = tfds.load(
    name="imdb_reviews",
    split=('train[:80%]', 'train[20%:]', 'test'),
    as_supervised=True)
```

```python
def preprocess_tfds(dataset):
    texts = []
    labels = []
    for text, label in tfds.as_numpy(dataset):
        texts.append(str(text, 'utf-8'))
        labels.append(label)
    return np.array(texts), np.array(labels)

train_texts, train_labels = preprocess_tfds(train_data)
val_texts, val_labels = preprocess_tfds(validation_data)
test_texts, test_labels = preprocess_tfds(test_data)

def preprocess_text(text):
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    return text

train_texts = [preprocess_text(text) for text in train_texts]
val_texts = [preprocess_text(text) for text in val_texts]
test_texts = [preprocess_text(text) for text in test_texts]
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=5000)
X_train = vectorizer.fit_transform(train_texts)
X_val = vectorizer.transform(val_texts)
X_test = vectorizer.transform(test_texts)
```

```python
from sklearn.svm import SVC # Import SVC explicitly

nb_model = MultinomialNB()
nb_model.fit(X_train, train_labels)
nb_preds = nb_model.predict(X_test)
print("Naive Bayes Accuracy:", accuracy_score(test_labels, nb_preds))

svm_model = SVC(kernel='linear') # Now SVC is recognized
svm_model.fit(X_train, train_labels)
svm_preds = svm_model.predict(X_test)
print("SVM Accuracy:", accuracy_score(test_labels, svm_preds))
```

```python
# Unigram BoW
X_train_unigram = unigram_vectorizer.fit_transform(train_texts)
X_test_unigram = unigram_vectorizer.transform(test_texts)

# Bigram BoW
X_train_bigram = bigram_vectorizer.fit_transform(train_texts)
X_test_bigram = bigram_vectorizer.transform(test_texts)

# Trigram BoW
X_train_trigram = trigram_vectorizer.fit_transform(train_texts)
X_test_trigram = trigram_vectorizer.transform(test_texts)

# Unigram TF-IDF
X_train_unigram_tfidf = unigram_tfidf.fit_transform(train_texts)
X_test_unigram_tfidf = unigram_tfidf.transform(test_texts)

# Bigram TF-IDF
X_train_bigram_tfidf = bigram_tfidf.fit_transform(train_texts)
X_test_bigram_tfidf = bigram_tfidf.transform(test_texts)

# Trigram TF-IDF
X_train_trigram_tfidf = trigram_tfidf.fit_transform(train_texts)
X_test_trigram_tfidf = trigram_tfidf.transform(test_texts)
```

```python
def train_and_evaluate(model, X_train, y_train, X_test, y_test, label):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    acc = accuracy_score(y_test, predictions)
    print(f"{label} Accuracy: {acc:.4f}")

nb_model = MultinomialNB()

print("Bag of Words (BoW) with Naive Bayes:")
train_and_evaluate(nb_model, X_train_unigram, train_labels, X_test_unigram, test_labels, "Unigram BoW")
train_and_evaluate(nb_model, X_train_bigram, train_labels, X_test_bigram, test_labels, "Bigram BoW")
train_and_evaluate(nb_model, X_train_trigram, train_labels, X_test_trigram, test_labels, "Trigram BoW")

print("\nTF-IDF with Naive Bayes:")
train_and_evaluate(nb_model, X_train_unigram_tfidf, train_labels, X_test_unigram_tfidf, test_labels, "Unigram TF-IDF")
train_and_evaluate(nb_model, X_train_bigram_tfidf, train_labels, X_test_bigram_tfidf, test_labels, "Bigram TF-IDF")
train_and_evaluate(nb_model, X_train_trigram_tfidf, train_labels, X_test_trigram_tfidf, test_labels, "Trigram TF-IDF")
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional
import tensorflow_datasets as tfds
import numpy as np
import re

max_vocab_size = 10000  # Maximum vocabulary size
max_len = 128           # Maximum sequence length

train_data, validation_data, test_data = tfds.load(
    name="imdb_reviews",
    split=('train[:80%]', 'train[20%:]', 'test'),
    as_supervised=True)

def preprocess_tfds(dataset):
    texts = []
    labels = []
    for text, label in tfds.as_numpy(dataset):
        texts.append(str(text, 'utf-8'))
        labels.append(label)
    return np.array(texts), np.array(labels)

train_texts, train_labels = preprocess_tfds(train_data)
val_texts, val_labels = preprocess_tfds(validation_data)
test_texts, test_labels = preprocess_tfds(test_data)

def preprocess_text(text):
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    return text

train_texts = [preprocess_text(text) for text in train_texts]
val_texts = [preprocess_text(text) for text in val_texts]
test_texts = [preprocess_text(text) for text in test_texts]

tokenizer = Tokenizer(num_words=max_vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(train_texts)


X_train_seq = tokenizer.texts_to_sequences(train_texts)
X_val_seq = tokenizer.texts_to_sequences(val_texts)
X_test_seq = tokenizer.texts_to_sequences(test_texts)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding="post", truncating="post")
X_val_pad = pad_sequences(X_val_seq, maxlen=max_len, padding="post", truncating="post")
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding="post", truncating="post")

embedding_dim = 128
lstm_units = 64

lstm_model = Sequential([
    Embedding(input_dim=max_vocab_size, output_dim=embedding_dim, input_length=max_len),
    Bidirectional(LSTM(units=lstm_units, return_sequences=False)),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # Binary classification
])

lstm_model.compile(optimizer='adam',
                   loss='binary_crossentropy',
                   metrics=['accuracy'])
# Train the model
lstm_history = lstm_model.fit(
    X_train_pad, train_labels,
    validation_data=(X_val_pad, val_labels),
    epochs=5,
    batch_size=32
)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


lstm_loss, lstm_accuracy = lstm_model.evaluate(X_test_pad, test_labels, verbose=0)
lstm_predictions = (lstm_model.predict(X_test_pad) > 0.5).astype("int32")  # Convert probabilities to binary predictions

lstm_precision = precision_score(test_labels, lstm_predictions)
lstm_recall = recall_score(test_labels, lstm_predictions)
lstm_f1 = f1_score(test_labels, lstm_predictions)

# ... (Rest of your code for comparison and plotting) ...

comparison = {
    "Metric": ["Accuracy", "Precision", "Recall", "F1-Score"],
    "LSTM": [lstm_accuracy, lstm_precision, lstm_recall, lstm_f1],
    "BERT": [0.8777, 0.9082, 0.8402, 0.8729]
}

comparison_df = pd.DataFrame(comparison)
print(comparison_df)
```

```python
def predict_sentiment(text):
    # Tokenize and pad the input text
    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=max_len, padding="post", truncating="post")
    prediction = (lstm_model.predict(padded) > 0.5).astype(int)
    return "Positive" if prediction[0] == 1 else "Negative"

print(predict_sentiment("The movie was fantastic and very engaging!"))
print(predict_sentiment("I did not like the movie. It was boring and slow."))
```
